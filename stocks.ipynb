## Task-9
## What was the correlation between the daily returns of different stocks?

## correlation by scatter_matrix

```python
from pandas.plotting import scatter_matrix
```

```python
a=pd.DataFrame(daily_return_avg_of_apple)
b=pd.DataFrame(daily_return_avg_of_google)
c=pd.DataFrame(daily_return_avg_of_microsoft)
d=pd.DataFrame(daily_return_avg_of_amazon)
```

```python
z=pd.concat([a['close'],b['close'],c['close'],d['close']],axis=1)
z.columns=['Apple_Daily_Returns','Google_Daily_Returns','Microsoft_Daily_Returns','Amazon_Daily_Returns']
z
```

```python
scatter_matrix(z,figsize=(20,10),hist_kwds={'bins':50})
plt.show()
```

## correlation by heatmap

```python
import numpy as np
m=np.ones_like(z.corr())
m[np.tril_indices_from(m)]=0
m
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(z.corr(),annot=True,mask=m,cmap='Blues')
plt.show()
```
